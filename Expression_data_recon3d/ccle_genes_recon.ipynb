```python
# https://depmap.org/portal/data_page/?tab=allData - version22Q2 downloaded
import pandas as pd
csv = pd.read_csv(r'/Users/subasrees/Downloads/CCLE_expression.csv')
csv.head()
```

```python
df_csv = pd.DataFrame(data=csv)
df = pd.DataFrame(df_csv.T)
df.columns = df.iloc[0]
df = df[1:]
df.head()
```

```python
df['Genes']=df.index
df.head()
```

```python
# get two seperate columns for gene symbols and entrez ids
df[['Symbol','Entrez']] = df['Genes'].apply(
   lambda x: pd.Series(str(x).split("(")))
df[['Entreznew','Dummy']] = df['Entrez'].apply(
   lambda x: pd.Series(str(x).split(")")))   
```

```python
# we are using entrz ids to extract recon 3d ids, remove the unwanted columns
cols_remove=['Genes','Symbol','Entrez','Dummy']
df.drop(cols_remove,axis=1,inplace=True)
```

```python
# load an excel file which relates commonly used gene ids with all the genes in recon 3d
x_source=pd.read_excel('/Users/subasrees/Desktop/python data/comparison.xlsx')
x_source.reset_index(inplace=True)
```

```python
## Use only the columns entrez ids (entrez_id) and recon ids (gene_number)
x_refer_ccle=x_source[['gene_number','entrez_id']]
#x_refer_ccle=x_source.groupby(['ensemble_id','gene_number']).count().reset_index()
#x_refer_ccle=x_refer_ccle[['ensemble_id','gene_number']]
x_refer_ccle['entrez_id']=x_refer_ccle['entrez_id'].astype(int)
#x_refer_ccle['ensemble_id']=x_refer_ccle['ensemble_id']
x_refer_ccle=x_refer_ccle.dropna()
x_refernew_ccle=x_refer_ccle.drop_duplicates(subset=['gene_number'],keep = 'first')
x_refernew_ccle.head()
```

```python
## rename the column name in ccle data to 'entrez_id'
df['entrez_id']=df['Entreznew']
full_ccle=df.dropna()
full_ccle=full_ccle.drop_duplicates(subset=['entrez_id'],keep = 'first')
full_ccle.reset_index(drop=True,inplace=True)
full_ccle['entrez_id']=full_ccle['entrez_id'].astype(int)
```

```python
#to_ccle=pd.merge(x_refernew_ccle,x_refer_ccle,how="left",on='ensemble_id')
to_ccle=pd.merge(full_ccle,x_refernew_ccle,how="inner",on='entrez_id')
```

```python
# recon ids extracted to the ccle data
recon_ccle=to_ccle.dropna()
recon_ccle.head()
```

```python
df_final=pd.DataFrame(recon_ccle)
rows = len(df_final.axes[0])
cols = len(df_final.axes[1])

# Print the number of rows and columns
print("Number of Rows: " + str(rows))
print("Number of Columns: " + str(cols))
```
