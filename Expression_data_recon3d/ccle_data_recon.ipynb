```python
## ccle https://depmap.org/portal/data_page/?tab=allData
import pandas as pd
csv = pd.read_csv(r'/Users/subasrees/Downloads/CCLE_expression.csv')
csv.head()
```

```python
df_csv = pd.DataFrame(data=csv)
df = pd.DataFrame(df_csv.T)
df.columns = df.iloc[0]
df = df[1:]
df.head()
```

```python
df['Genes']=df.index
df.head()
```

```python
## ccle
df[['Symbol','Entrez']] = df['Genes'].apply(
   lambda x: pd.Series(str(x).split("(")))
df[['Entreznew','Dummy']] = df['Entrez'].apply(
   lambda x: pd.Series(str(x).split(")")))   
```

```python
#ccle
cols_remove=['Genes','Symbol','Entrez','Dummy']
df.drop(cols_remove,axis=1,inplace=True)
full_x_ccle=df
```

```python
# create a dataframe with entrez id columns
x_refer_ccle=df['Entreznew'].dropna() #ccle
x_refer_ccle=pd.DataFrame(x_refer_ccle)
x_refernew_ccle=x_refer_ccle.drop_duplicates(keep = 'first')
x_refernew_ccle.head()
```

```python
# load an excel which relates entrez id with recon 3d ids
x_source=pd.read_excel('/Users/subasrees/Desktop/python data/comparison.xlsx')
x_source.head()
```

```python
## First extract the recon ids present in the column gene_number to the entrez ids present in the data
main_x_ccle=x_source.groupby(['entrez_id','gene_number']).count().reset_index()
#main_x_ccle=x_source.groupby(['ensemble_id','gene_number']).count().reset_index()
#main_x_ccle=main_x_ccle[['ensemble_id','gene_number']]
main_x_ccle=main_x_ccle[['entrez_id','gene_number']]
main_x_ccle=pd.DataFrame(main_x_ccle)
main_x_ccle['entrez_id']=main_x_ccle['entrez_id'].astype(int)
#main_x_ccle['ensemble_id']=main_x_ccle['ensemble_id']
main_x_ccle.head()
```

```python
#ccle
#x_refernew_ccle.columns=['ensemble_id']
x_refernew_ccle.columns=['entrez_id']
#x_refernew_ccle.columns=['ensemble_id']
x_refernew_ccle['entrez_id']=x_refernew_ccle['entrez_id'].astype(int)
#x_refernew_ccle['ensemble_id']=x_refernew_ccle['ensemble_id']
x_refernew_ccle=pd.DataFrame(x_refernew_ccle)
x_refernew_ccle.head()
```

```python
#to_ccle=pd.merge(x_refernew_ccle,main_x_ccle,how="left",on='ensemble_id')
to_ccle=pd.merge(x_refernew_ccle,main_x_ccle,how="left",on='entrez_id')
```

```python
# recon ids extracted
newtop_ccle=to_ccle.dropna()
newtop_ccle.head()
```

```python
gene_number_id_ccle=pd.DataFrame(newtop_ccle)
rows = len(gene_number_id_ccle.axes[0])
cols = len(gene_number_id_ccle.axes[1])

# Print the number of rows and columns
print("Number of Rows: " + str(rows))
print("Number of Columns: " + str(cols))
##
gene_number_id_ccle = gene_number_id_ccle.set_axis(["Entreznew", "gene_number"], axis=1)
columnid_ccle=gene_number_id_ccle[["Entreznew"]]
gene_number_id_ccle.head()
```

```python

## change the entrez ids to integer types
gene_number_id_ccle['Entreznew']=gene_number_id_ccle['Entreznew'].astype(int)
full_x_ccle['Entreznew']=full_x_ccle['Entreznew'].astype(int)
```

```python

df_final=pd.merge(gene_number_id_ccle,full_x_ccle,on='Entreznew')
```

```python
# Get the number of rows and columns
rows = len(df_final.axes[0])
cols = len(df_final.axes[1])
  
# Print the number of rows and columns
print("Number of Rows: " + str(rows))
print("Number of Columns: " + str(cols))
```
